## Exploratory Data Analysis


### Research Question 1 Which movies are the most popular of all times?

```python
df['profit']=df['revenue']-df['budget']
```

```python
def minmax(x):
    mins=df[x].idxmin()
    maax=df[x].idxmax()
    high=pd.DataFrame(df.loc[maax,:])
    low=pd.DataFrame(df.loc[mins,:])
    
    print("Movie has the highest "+x+ ":",df['original_title'][maax])
    print("Movie has the lowest "+x+ ":",df['original_title'][mins])
    return pd.concat([high,low],axis=1)

minmax('profit')
```

- The first column shows the highest profitable movie and second column shows the highest in loss movie of the dataset.
- Avatar movie directed by james cameron has 7.1 vote average with good tagline.

### Research Question 2  Which movies have the highest and lowest budget?

```python
df['budget']=df['budget']
minmax('budget')
```

```python
info=pd.DataFrame(df['budget'].sort_values(ascending=False))
info['original_title']=df['original_title']
data=list(map(str,(info['original_title'])))

x=list(data[:10])
y=list(info['budget'][:10])

ax=sns.pointplot(y,x)
sns.set(rc={'figure.figsize':(10,5)})
ax.set_title("Top 10 High Budget Movies",fontsize = 15)
ax.set_xlabel("Budget",fontsize = 13)
sns.set_style("darkgrid");
```

- The one with highest budget has many casts and different genre which can be cause of more budgeted movie with very less profit.
- This clearly shows the lowest budget movie and highest budget movie.

### Research Question 3 which year has the highest released movie?

```python
d1=df.groupby('release_year').count()['id']
print(d1.tail())

df.groupby('release_year').count()['id'].plot(xticks=np.arange(1960,2016,5))

sns.set(rc={'figure.figsize':(10,6)})
plt.title("Years vs Number of movies",fontsize=12)
plt.xlabel('Release year',fontsize=10)
plt.ylabel('Number of Movies',fontsize=10)
sns.set_style("whitegrid");
```

- This graph represents the released movie in different year. As there are many years, gap interval of 5 is given so that there is no clutter in graph.
