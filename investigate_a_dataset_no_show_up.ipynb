### Research Question 1: Will the demographics have the impact on the ratio of showing up for appointment?

```python
rs1_data = filtered_data.drop([ 'ScheduledDay', 'AppointmentDay', 'Scholarship', 'Hipertension',
       'Diabetes', 'Alcoholism', 'Handcap', 'SMS_received'], axis = 1)
```

```python
rs1_data.head()
```

```python
no_show = rs1_data.no_show == 'No'
show = rs1_data.no_show == 'Yes'
```

#### Gender
Women are more likely to show up and not show up at the appointment


```python
rs1_data.groupby(['no_show','Gender']).count()
```

```python
rs1_data.groupby(['no_show','Gender']).count()/rs1_data.count()*100
```

#### Age

As people age, they are less likely to show up for the appointment. While among those who don't show up for the appointments, people who aged less than 10 and between 45 to 55 are those who are more likely to skip the appointments

```python
rs1_data.Age[no_show].hist(alpha=0.5, bins=10, label='No Show Up')
rs1_data.Age[show].hist(alpha=0.5, bins=20, label='Show')
plt.xlabel('Age')
plt.legend();
```

#### Neighbourhood

For each neighbourhood, the ratio of people who show up and not show up at the appointment is quite similar. Except for the Ilhas Oceaniscas de Trindade has 100% people showed up at the appointment.


```python
no_show_data = rs1_data.query('no_show == "No"')
```

```python
show_data = rs1_data.query('no_show == "Yes"')
```

```python
no_show_data.head()
```

```python
plt.figure(figsize=(18,6))  # Set the figure size
no_show_data.Neighbourhood.hist(alpha = 0.5, label = 'No Show Up')
show_data.Neighbourhood.hist(alpha = 0.5, label = "Show Up")
# Rotate x-axis tick labels
plt.xticks(rotation='vertical')
plt.xlabel('Neighborhood')
plt.legend()
plt.show()
```

### Research Question 2  Will people who have diseases or scholarship or receive SMS are more likely to show up for appointment?

People who received SMS more likely to show up at the meeting than people who have diseases or scholarship. people who have hypertension will more likely to not show up at the appointment



```python
rs2_data = filtered_data.drop(['Gender', 'Age', 'Neighbourhood', 'ScheduledDay','AppointmentDay', 'Handcap'], axis = 1)
rs2_data.head()
```

```python
rs2_data.rename(columns={'No-show': 'no_show'}, inplace = True)
```

```python
total = rs2_data['no_show'].count().astype(int)
```

```python
rs2_data.groupby(['no_show']).sum()
```

```python
ratio_table =rs2_data.groupby(['no_show']).sum()/total*100
ratio_table
```

```python
# Calculate the sum of two rows and create a new row
new_row = pd.DataFrame({'no_show': ['Sum'], 'Scholarship': [ratio_table['Scholarship'].sum()], 'Hipertension': [ratio_table['Hipertension'].sum()],
                        'Diabetes': [ratio_table['Diabetes'].sum()], 'Alcoholism': [ratio_table['Alcoholism'].sum()],
                         'SMS_received': [ratio_table['SMS_received'].sum()]})

new_row
```

```python
# Append the new row to the DataFrame
ratio_table = ratio_table.append(new_row, ignore_index = True)
ratio_table
```

```python
ratio_table.loc[0, 'no_show'] = 'No'
ratio_table.loc[1, 'no_show'] = 'Yes'
```

```python
ratio_table
```

```python

ratio_table.set_index('no_show', inplace=True)
ratio_table = ratio_table.round(1)
ratio_table = ratio_table.T

# set no_show column as a index
ax = ratio_table.plot(kind='bar', figsize=(10, 6))

# Set labels and title
plt.xlabel('Ratio of Showing Up for each disease')
plt.ylabel('Ratio')
plt.title('Ratio of Showing Up for Disease')
plt.grid(True)
plt.legend()
plt.show()
```
