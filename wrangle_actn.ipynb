 ##### 2- Create list of prediction confidence and list of prediction_images .

##### # Define : create list of prediction confidence as (p_con) and list of prediction_images as (pred)

##### # Code :

```python
image_predictions_copy
```

```python
image_predictions_copy['p1_dog'].value_counts()
```

```python
image_predictions_copy['p2_dog'].value_counts()
```

```python
image_predictions_copy['p3_dog'].value_counts()
```

```python

pred = []
# confidence list
p_conf = []


# function to check the value if True will collected :

def dog(image_predictions_copy):
    
    if image_predictions_copy['p1_dog'] == True:
        pred.append(image_predictions_copy['p1'])
        p_conf.append(image_predictions_copy['p1_conf'])
    elif image_predictions_copy['p2_dog'] == True:
        pred.append(image_predictions_copy['p2'])
        p_conf.append(image_predictions_copy['p2_conf'])
    elif image_predictions_copy['p3_dog'] == True:
        pred.append(image_predictions_copy['p3'])
        p_conf.append(image_predictions_copy['p3_conf'])
    else:
        pred.append('None')
        p_conf.append('None')

        
image_predictions_copy.apply(dog, axis=1)

# create new columns  :

image_predictions_copy['pred'] = pred
image_predictions_copy['p_conf'] = p_conf
```

```python

# drop rows that have None :
image_predictions_copy = image_predictions_copy[image_predictions_copy['pred'] != 'None']

```
