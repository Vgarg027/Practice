```python
df_mov=df_mov.drop(['cast', 'homepage', 'tagline', 'keywords', 'overview', 'imdb_id'], axis=1)
```

```python
df_mov['budget'] = df_mov['budget'].replace(0,df_mov['budget'].mean())

df_mov['revenue'] = df_mov['revenue'].replace(0,df_mov['revenue'].mean())
```

```python
df_mov['genres'] = df_mov['genres'].apply(lambda x: x.split('|')[0])
df_mov['production_companies'] = df_mov['production_companies'].apply(lambda x: x.split('|')[0])
```

```python
df_mov
```

```python
genres_popularity = df_mov.groupby(['genres'])['popularity'].mean()
print (genres_popularity.index)
```

```python
plt.subplots(figsize=(20, 6))
plt.bar(genres_popularity.index, genres_popularity)
plt.title('popularity by genre')
plt.xlabel('genre')
plt.ylabel('Popularity');
```
