### Converting is Done

```python
ordinal_var_dict = {'ProsperRating (Alpha)': ['N/A','HR','E','D','C', 'B', 'A', 'AA'],
                    'IncomeRange':['$0','$1-24,999','$25,000-49,999','$50,000-74,999','$75,000-99,999','$100,000+']}
for var in ordinal_var_dict:
    ordered_var = pd.api.types.CategoricalDtype(ordered = True, categories = ordinal_var_dict[var])
    prosper_loan[var] = prosper_loan[var].astype(ordered_var)
```

```python
#Adjust datatype for all other categorical columns
prosper_loan['Occupation'] = prosper_loan['Occupation'].astype('category')
prosper_loan['BorrowerState'] = prosper_loan['BorrowerState'].astype('category')
prosper_loan['LoanStatus'] = prosper_loan['LoanStatus'].astype('category')
prosper_loan['ListingCategory (numeric)'] = prosper_loan['ListingCategory (numeric)'].astype('category')
prosper_loan['EmploymentStatus'] = prosper_loan['EmploymentStatus'].astype('category')
```

```python
base_color = sb.color_palette()[0]
n_employment = prosper_loan.EmploymentStatus.value_counts().sum()
max_n_employment = prosper_loan.EmploymentStatus.value_counts()[0]
max_prop_employment = max_n_employment / n_employment
employment_ticks = np.arange(0,max_prop_employment,0.1)
tick_names = ['{:0.2f}'.format(v) for v in employment_ticks]
```
