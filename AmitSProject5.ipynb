## Data Wrangling

> **Phase1**:  It involves a combination of manual processing and automation using tools and techniques such as filtering, aggregation, and merging data from multiple sources.

Data wrangling is a crucial step in the data analysis process, as it ensures that the data is in a consistent and usable form. It can be a time-consuming and tedious task, but it is essential for producing reliable and accurate results.

There are several steps involved in data wrangling, including:

Acquiring the data: This involves collecting the data from various sources such as databases, APIs, or flat files.

Cleaning the data: This involves removing or correcting any errors, inconsistencies, or missing values in the data.

Structuring the data: This involves organizing the data into a consistent and logical format, such as a table or a database.

Transforming the data: This involves changing the data into a form that is more suitable for analysis, such as converting data types or creating new variables.

Validating the data: This involves checking the data to ensure that it is accurate and complete.

Data wrangling is a crucial and often underappreciated aspect of data analysis. By investing the time and effort to properly wrangle your data, you can ensure that your analysis is based on reliable and accurate data, and that your findings are meaningful and useful.

#### Remove Duplicates

```python
# check duplicated rows
print('There are {} Duplicated Rows'.format(movie_df.duplicated().sum()))
```

```python
# remove duplicated rows
movie_df.drop_duplicates(inplace=True)
print("After removing duplicated rows the number of rows become {} ".format(movie_df.shape[0]))
```

#### Drop Unused Columns:
*I will drop unused columns like 'imdb_id', 'cast', 'homepage', 'tagline', 'keywords', 'overview', 'production_companies', 'vote_count', 'budget_adj' and 'revenue_adj' columns from data as these columns are not useful for my analysis*

```python
movie_df.drop(columns=['imdb_id',  'cast', 'homepage', 'tagline', 'keywords', 'overview', 'production_companies', 'vote_count', 'budget_adj', 'revenue_adj'], inplace=True)
print("After Removing Unused Columns There are {} Columns,".format(movie_df.shape[1]))
print('and their Names are')
print(movie_df.columns.values)
```

#### Remove Null Values:


```python
# check count of null values
movie_df.isnull().sum()

#1) 'director' column have 44 null values.
#2) 'genres' column have 23 null values.
```

##### __ 
#1) 'director' column have 44 null values.
#2) 'genres' column have 23 null values. __

```python
# remove null values
movie_df.dropna(inplace=True)
```

```python
# check count of null values
movie_df.isnull().sum()
```

```python
# print number of rows and columns
print("There are {} Rows and {} Columns After Removing Null Values".format(movie_df.shape[0], movie_df.shape[1]))
```

#### Fix Zero Values in Budget, Revenue and Runtime:

```python
# check zero values in 'budget', 'revenue' and 'runtime' columns.
movie_df[['budget', 'revenue', 'runtime']].apply(lambda x: np.sum(x==0))
```

##### Observations:
1) 'budget' column has 5636 zero values.

2) 'revenue' column has 5952 zero values.

3) 'runtime' column has 28 zero values.

```python
# replace zero values with the mean of each column
movie_df[['budget', 'revenue', 'runtime']] = movie_df[['budget', 'revenue', 'runtime']].apply(lambda x: x.replace(0, x.mean()))
```

```python
# check zero values
movie_df[['budget', 'revenue', 'runtime']].apply(lambda x: np.sum(x==0))
```

### Fix Columns Datatypes:


```python
# check columns datatypes.
movie_df.dtypes
```

#### Comments: 

1) I will convert 'id' column from integer to string.

2) I will convert 'release_date' column from object to datetime.

```python
# convert 'id' column datatype to string format
movie_df['id'] = movie_df['id'].astype('string')
```

```python
# convert 'release_date' column datatype to datetime format
movie_df['release_date'] = pd.to_datetime(movie_df['release_date'])
```

```python
# check columns datatypes.
movie_df.dtypes
```

#### Add New Column & rename others :


```python
# add 'profit' column
movie_df['profit'] = movie_df['revenue'] - movie_df['budget']
```

```python
# rename 'genres' to 'genre', 'original_title' to 'title', 'vote_average' to 'rating', 'release_year' to 'year'
movie_df.rename(columns={'genres':'genre', 'original_title':'title', 'vote_average':'rating', 'release_year':'year'}, inplace=True)
print("Current Names are {}".format(movie_df.columns.values))
```

```python
# rearrange columns positions
movie_df = movie_df[['id', 'title', 'director', 'genre', 'runtime', 'budget', 'revenue', 'profit', 'rating',  'popularity', 'release_date', 'year']]
print("Current Order is {}".format(movie_df.columns.values))
```

### Fix Genre Values


```python
def explode_values(df,col):
    """ This Function Take dataframe 'df' and column 'col' with multiple values and return exploded dataframe"""
    # split column values into lists
    df[col] = df[col].str.split('|')
    # return exploded dataframe
    return df.explode(col, ignore_index=True)
    
```

```python
# explode 'genre' values 
movie_df = explode_values(movie_df, 'genre')
```

```python
# check 'genre' values
print("Genre Values became ", movie_df.genre.unique())
```

```python
# check 'genre' values
print("Genre Values became ", movie_df.genre.unique())
```
