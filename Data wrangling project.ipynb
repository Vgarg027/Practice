# Cleaning Data
creating data copies

```python
clean_t_archive = t_archive.copy()
clean_df_image = df_image.copy()
clean_df_tweet = df_tweet.copy().rename(columns = {'id':'tweet_id'})
```

```python
#rename id co;umn to tweet_id in clean_df_tweet
clean_df_tweet.columns
```

```python
# Cleaning tidiness 
#dog stages(i.e doggo,floofer,pupper and puppo)should be in one column called
#(dog_stages) 
clean_t_archive['dog_stages'] = clean_t_archive['text'].str.extract('(doggo|floofer|pupper|puppo)')
```

```python
#remove unbenefit columns.
clean_t_archive.drop(['doggo','floofer','pupper','puppo'],axis = 1, inplace = True)
```

```python
clean_t_archive.columns
```

```python
clean_df_image.columns
```

```python
clean_df_tweet.columns
```

```python
clean_t_archive.columns
```

```python
# Combine DataFrames with left join.
clean_t_archive = pd.merge(clean_t_archive,clean_df_tweet,on = 'tweet_id',how = 'left')
clean_t_archive = pd.merge(clean_t_archive,clean_df_image,on = 'tweet_id',how = 'left')
```

```python
clean_t_archive.shape
```
